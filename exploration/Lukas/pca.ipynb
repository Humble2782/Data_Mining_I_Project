# Factor Analysis of Mixed Data (FAMD)

FAMD is a dimensionality reduction technique that combines the principles of:

- **Principal Component Analysis (PCA)** — for continuous numerical variables  
- **Multiple Correspondence Analysis (MCA)** — for categorical variables  

By integrating both, FAMD provides a unified low-dimensional representation of mixed-type data.  
It captures underlying structures, correlations, and groupings that may not be visible in the original feature space.

### Objectives
- Preprocess a dataset containing both numeric and categorical features  
- Apply **FAMD** using the `prince` library in Python  
- Examine how variables contribute to the main dimensions  
- Visualize the resulting component space to identify patterns and relationships  

### Why FAMD?
Unlike PCA (which assumes all features are numeric), FAMD:
- Handles categorical and numeric features simultaneously  
- Balances their influence automatically (without one dominating the analysis)  
- Enables interpretation of complex datasets with mixed data types  

### Step 1: Differentiate between numerical and categorical columns

````python
num_cols = ["speed_limit", "age", ]
cat_cols = []
for col in df_final.columns:
    if col not in num_cols:
        cat_cols.append(col)

df_final[cat_cols] = df_final[cat_cols].astype('category')
````

### Step 2: Initialize and Fit the FAMD Model

In this step, an instance of the **FAMD** model is created via the `prince` library and fitted to the dataset `df_final`.

Parameters:
- **`n_components=10`**: number of principal dimensions to extract  
- **`n_iter=3`**: number of iterations for the algorithm to converge  
- **`random_state=42`**: ensures reproducible results  
- **`engine="sklearn"`**: uses scikit-learn–compatible numerical routines  
- **`handle_unknown="error"`**: raises an error if unseen categorical levels appear during fitting  

Fitting the model (`famd.fit(df_final)`) computes the principal dimensions that capture the maximum joint variance from both numerical and categorical variables.

````python
famd = prince.FAMD(
    n_components=10,
    n_iter=10,
    copy=True,
    check_input=True,
    random_state=42,
    engine="sklearn",
    handle_unknown="error"
)

famd = famd.fit(df_final)
````

### Step 3: Examine Eigenvalues Summary

After fitting the FAMD model, the **eigenvalues summary** is checked using
`famd.eigenvalues_summary`.

The table shows:
- The **eigenvalues** for each extracted component (dimension)  
- The **percentage of inertia** (variance) explained by each component  
- The **cumulative percentage**, indicating how much total variance is captured as we include more dimensions  

Analyzing these values helps determine how many components are sufficient to represent most of the information in the dataset without significant loss.

````python
famd.eigenvalues_summary
````

### Step 4: Compute Row Coordinates

The command `famd.row_coordinates(df_final)` projects each observation (row) of the dataset into the **FAMD component space**.

Each row in the resulting table represents an observation’s position along the extracted dimensions.  
These coordinates summarize the original mixed variables into a few synthetic features that capture most of the dataset’s variability.

````python
famd.row_coordinates(df_final)
````

### Step 5: Inspect Column Coordinates

The attribute `famd.column_coordinates_` provides the **coordinates of variables** (both numerical and categorical) in the FAMD component space.

These coordinates indicate **how each variable contributes** to the extracted dimensions:
- Variables positioned farther from the origin have a **stronger influence** on those components.  
- Variables that lie close together are **positively correlated**, while those in opposite directions are **negatively correlated**.

````python
famd.column_coordinates_
````

### Step 6: Analyze Column Contributions

The attribute `famd.column_contributions_` shows the **relative contribution of each variable** to the formation of each FAMD component (dimension).

These contributions help understanding:
- **Which variables** are most influential in defining each component  
- How much **each feature contributes** to the total explained variance  
- Which dimensions capture the strongest **patterns of variation** in the dataset  

A higher contribution value means that the variable plays a larger role in shaping that particular component.

````python
famd.column_contributions_
````

### Final Step: Visualize Individuals in the FAMD Component Space

The plot displays the **projection of observations** (rows) from the dataset onto the first two FAMD components.

````python
famd.plot(
    df_final.head(1000),
    x_component=0,
    y_component=1,
)
````

### Cluster Observations with DBSCAN

After computing the FAMD components for the dataset, **clustering** is performed on the transformed observations to discover groups with similar profiles.

Used algorithm: **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**:

- **`eps`**: maximum distance between two points to consider them neighbors. Smaller values create more, tighter clusters.  
- **`min_samples`**: minimum number of points required to form a cluster. Lower values allow smaller clusters to form.  
- **Output**: each observation receives a **cluster label**. Noise points that do not belong to any cluster are labeled `-1`.

```python
# Example function to cluster observations
cluster_labels = cluster_famd(row_coords[[0,1]], eps=0.5, min_samples=5)
df_plot['DBSCAN_cluster'] = cluster_labels

````python
from sklearn.cluster import DBSCAN

def cluster_famd(row_coords, eps=0.1, min_samples=2):
    db = DBSCAN(eps=eps, min_samples=min_samples)
    db.fit(row_coords)
    return db.labels_.tolist()
````

Plotly version of the plot. Can be color-coded by parameter and cluster.

````python
import ipywidgets as widgets
from IPython.display import display
import plotly.express as px

# Compute clusters
df_plot['DBSCAN_cluster'] = cluster_famd(row_coords[[0,1]], eps=0.5, min_samples=5)
print(df_plot['DBSCAN_cluster'])

# Dropdown including original columns + cluster labels
color_dropdown = widgets.Dropdown(
    options=list(df_final.columns) + ['DBSCAN_cluster'],
    value=df_final.columns[0],
    description='Color by:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='300px')
)

def update_plot(color_col):
    fig = px.scatter(
        df_plot.head(1000),
        x=0,
        y=1,
        color=color_col,
        title=f"FAMD Projection (Colored by {color_col})",
        labels={"0": "Component 1", "1": "Component 2"},
        hover_data=df_final.columns,
        opacity=0.75,
    )
    fig.update_traces(marker=dict(size=6, line=dict(width=0.5, color='DarkSlateGrey')))
    fig.update_layout(template="plotly_white", width=800, height=600, title_x=0.5)
    fig.show()

widgets.interact(update_plot, color_col=color_dropdown)
````
